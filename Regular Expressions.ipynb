```python
text_to_search = '''
abcdefghijklmnopqrstuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
0123456789
1234567890
qwertyuiopasdfghjklzxcvbnm
QWERTYUIOPASDFGHJKLZXCVBNM
.
abcdefghijklmnopqrstuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
0123456789
1234567890
qwertyuiopasdfghjklzxcvbnm
QWERTYUIOPASDFGHJKLZXCVBNM
.
'''
```

```python
pattern = re.compile(r"abc")
matches = pattern.finditer(text_to_search)
counter = 0
for match in matches:
    print(match)
    print(match.span()[0], match.span()[1])
    counter += 1
print("Number of matches: ", counter)
```

```python
pattern = re.compile(".") # matches everyting except \n
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
sentence = "This is a random sentence"
```

```python
number = "123-123"
```

```python
pattern = re.compile(r"\d\d\d[-.]\d\d\d") # Square brackets means to find either of the included characters ex. - or .
matches = pattern.finditer(number)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r"[1-5]")
matches = pattern.finditer(number)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r"[1-5-]")
matches = pattern.finditer(number)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r"[A-z]")
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r"[^a-zA-Z]") # get everything except...
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r"\d{3}")
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r"1236?")
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r"\d+") # one or more
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r"1(2|3|4)")
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
emails = '''
omaremam@gmail.com
omare@hotmail.com
om@gmail.co
omar_emam@yahoo.com
omaoryahoo.com
omar@fue.edu
'''
```
