```python
text_to_search = '''
abcdefghijklmnopqrstuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
0123456789
1234567890
qwertyuiopasdfghjklzxcvbnm
QWERTYUIOPASDFGHJKLZXCVBNM
.
abcdefghijklmnopqrstuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
0123456789
1234567890
qwertyuiopasdfghjklzxcvbnm
QWERTYUIOPASDFGHJKLZXCVBNM
.
'''
```

```python
pattern = re.compile(r"abc")
matches = pattern.finditer(text_to_search)
counter = 0
for match in matches:
    print(match)
    print(match.span()[0], match.span()[1])
    counter += 1
print("Number of matches: ", counter)
```

```python
pattern = re.compile(".") # matches everyting except \n
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```

```python
pattern = re.compile(r"\.") # matches everyting except \n
matches = pattern.finditer(text_to_search)
for match in matches:
    print(match)
```
