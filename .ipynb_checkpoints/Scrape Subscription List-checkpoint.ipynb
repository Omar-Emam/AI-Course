```python
from bs4 import BeautifulSoup
import csv
```

```python
#Used a local html file as youtube link would as for login
with open("subscription_list.html", 'r', encoding="utf8") as html_file: 
    soup = BeautifulSoup(html_file, 'lxml')
print(soup.prettify())
```

```python
channel = soup.find('ytd-channel-renderer')
print(channel.prettify())
```

```python
channel_name = channel.find('yt-formatted-string', id='text').text
print(channel_name)
```

```python
channel_src = channel.find('a', id='main-link')['href']
channel_src
```

```python
channel_link = f'https://www.youtube.com{channel_src}'
channel_link
```

```python
for channel in soup.find_all('ytd-channel-renderer'):    
    channel_name = channel.find('yt-formatted-string', id='text').text
    print(channel_name)
    
    channel_src = channel.find('a', id='main-link')['href']
    channel_link = f'https://www.youtube.com{channel_src}'
    print(channel_link)
```
